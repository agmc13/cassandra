```python
import matplotlib.pyplot as plt
```

```python
class RDD:
    global df
    global df1
    def __init__(self): 
        global df1
        df = spark.read\
            .format("org.apache.spark.sql.cassandra")\
            .options(table="datosespania", keyspace="espania")\
            .load()
        df1 = df.toPandas()
        df.groupBy("estado").count().orderBy('estado', ascending=False).show()
    def grafica(self):
        df1["estado"].value_counts().plot(kind='bar', alpha=0.5)
```

```python
datos=RDD()
```

```python
datos.grafica()
```
