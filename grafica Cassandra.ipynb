```python
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession 
```

```python
class RDD:
    def __init__(self):
        global df1
        spark = SparkSession.builder.appName("Datos Cassandra").config("spark.some.config.option", "prueba").getOrCreate() 
        df = spark.read\
            .format("org.apache.spark.sql.cassandra")\
            .options(table="datosespania", keyspace="espania")\
            .load()
        df1 = df.toPandas()
        df.groupBy("estado").count().orderBy('estado', ascending=False).show()
    def grafica(self):
        plt.figure(figsize=(15,10))
        df1["estado"].value_counts().plot(kind='bar', alpha=1)
```

```python
datos=RDD()
```

```python
datos.grafica()
```
